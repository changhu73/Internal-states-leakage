```python
import matplotlib.pyplot as plt

# Data
models = ['1B', '3B', '8B', '70B']
input_only_acc = [0.7083, 0.7115, 0.7289, 0.9246]  # ACC data in decimals
ref_added_acc = [0.7051, 0.7372, 0.7711, 0.9541]
input_only_f1 = [0.7316, 0.7632, 0.7736, 0.9292]  # F1 data in decimals
ref_added_f1 = [0.7604, 0.7760, 0.7862, 0.9557]
x = range(len(models))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, input_only_acc, label='Input Only - ACC', marker='o', linestyle='-')
plt.plot(x, ref_added_acc, label='Ref Added - ACC', marker='o', linestyle='--')
plt.plot(x, input_only_f1, label='Input Only - F1', marker='s', linestyle='-')
plt.plot(x, ref_added_f1, label='Ref Added - F1', marker='s', linestyle='--')

# Annotating data points
for i in range(len(x)):
    plt.text(x[i], input_only_acc[i], f'{input_only_acc[i]:.4f}', ha='center', fontsize=9)
    plt.text(x[i], ref_added_acc[i], f'{ref_added_acc[i]:.4f}', ha='center', fontsize=9)
    plt.text(x[i], input_only_f1[i], f'{input_only_f1[i]:.4f}', ha='center', fontsize=9)
    plt.text(x[i], ref_added_f1[i], f'{ref_added_f1[i]:.4f}', ha='center', fontsize=9)

# Adding labels and title
plt.xticks(x, models, fontsize=12)  # Adjust x-tick font size
plt.yticks(fontsize=12)  # Adjust y-tick font size
plt.xlabel('Model Size', fontsize=14)  # Adjust x-axis label font size
plt.ylabel('Performance', fontsize=14)  # Adjust y-axis label font size
plt.title('Performance Trends with Increasing Model Size (Decimal Data)', fontsize=16)  # Adjust title size

# Customizing legend without a border
plt.legend(frameon=False)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['1B', '3B', '8B', '13B', '70B']
input_only_acc = [0.7083, 0.7115, 0.7289, 0.7883, 0.8524]  # ACC data in decimals
input_only_f1 = [0.7316, 0.7632, 0.7736, 0.7644, 0.8807]  # F1 data in decimals
ref_added_acc = [0.7051, 0.7372, 0.7711, 0.8003, 0.8729]
ref_added_f1 = [0.7604, 0.7760, 0.7862, 0.7915, 0.8994]

# Use a more appropriate x range for the model sizes
x = np.arange(len(models))  # X-axis positions for the models
bar_width = 0.2  # Width of each bar

# Plotting
plt.figure(figsize=(14, 10))  # Increase figure size further

# Plot bars for each metric
plt.bar(x - 1.5 * bar_width, input_only_acc, bar_width, label='Input Only - ACC', color='b', alpha=0.7)
plt.bar(x - 0.5 * bar_width, input_only_f1, bar_width, label='Input Only - F1', color='r', alpha=0.7)
plt.bar(x + 0.5 * bar_width, ref_added_acc, bar_width, label='Ref Added - ACC', color='g', alpha=0.7)
plt.bar(x + 1.5 * bar_width, ref_added_f1, bar_width, label='Ref Added - F1', color='c', alpha=0.7)

# Adding labels and title
plt.xticks(x, models, fontsize=24)  # Custom x-ticks with larger font size
plt.yticks(fontsize=24)  # Increase y-tick font size
plt.xlabel('Model Size', fontsize=28)  # Larger x-axis label font size
plt.ylabel('Performance', fontsize=28)  # Larger y-axis label font size

# Set the title at the bottom
# plt.suptitle('Effect of Model Size', fontsize=30, y=-0.01)  # Main title at the bottom

# Add grid
plt.grid(True, axis='y', linestyle='--', alpha=0.6)  # Add horizontal grid with dashed lines and transparency

# Customizing legend without a border
plt.legend(frameon=False, fontsize=24)  # Larger legend font size

# Truncate y-axis from 0.6 to 1.0
plt.ylim(0.65, 1.0)  # Set y-axis range from 0.6 to 1.0

# Add a tilde (~) at y=0.6 to indicate truncation
ax = plt.gca()  # Get current axis
# ax.axhline(0.65, color='black', linestyle='--', linewidth=1.5)  # Draw a horizontal line at y=0.6
ax.text(0.05, 0.66, '', fontsize=30, color='black', verticalalignment='bottom')  # Add tilde at y=0.6

# Display the plot
plt.tight_layout()
plt.show()
```
